## data
in this file we can visualize the motion data of one perticular frame

please **run asfamc2tensor.py first** to generate the coordinates of nodes and adjacency matrix

```python
adj = torch.load('adj09.pkl')
xs = torch.load('xs09.pkl')
```

## draw

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
H0 = xs[0,:,:]
N = adj.shape[0]
```

```python
fig = plt.figure(figsize=(6,6))
ax = Axes3D(fig)

ax.set_xlim3d(-0.2,0.6)
ax.set_ylim3d(-0.2,0.6)
ax.set_zlim3d(0,0.8)

xs, ys, zs = [], [], []
for i in range(H0.shape[0]):
    xs.append(H0[i,0])
    ys.append(H0[i,1])
    zs.append(H0[i,2])

plt.plot(zs, xs, ys, 'b.')

# for edge in edges:
for i in range(N):
    for j in range(N):
        if adj[i,j] == 1:
            xs = [H0[i,0],H0[j,0]]
            ys = [H0[i,1],H0[j,1]]
            zs = [H0[i,2],H0[j,2]]
            plt.plot(zs, xs, ys, 'r')

plt.show()
```
